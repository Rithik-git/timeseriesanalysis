```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

def check_stationarity(series, column_name):
    result = adfuller(series)
    adf_statistic = result[0]
    p_value = result[1]
    critical_values = result[4]

    # Interpretation
    is_stationary = p_value < 0.05

    print(f"\nColumn: {column_name}")
    print(f"ADF Statistic: {adf_statistic}")
    print(f"p-value: {p_value}")
    print("Critical Values:")
    for key, value in critical_values.items():
        print(f"   {key}: {value}")
    print(f"Is Stationary: {'Yes' if is_stationary else 'No'}")

def main():
    # Load dataset
    file_path = "/content/OzoneHole_Data.csv" # Update path if needed
    df = pd.read_csv(file_path)

    # Plot time series data
    plt.figure(figsize=(12,5))
    plt.subplot(2,1,1)
    plt.plot(df['Year'], df['Hole Area'], marker='o', linestyle='-')
    plt.title("Hole Area over Years")
    plt.xlabel("Year")
    plt.ylabel("Hole Area")

    plt.subplot(2,1,2)
    plt.plot(df['Year'], df['Minimum Ozone'], marker='o', linestyle='-', color='red')
    plt.title("Minimum Ozone over Years")
    plt.xlabel("Year")
    plt.ylabel("Minimum Ozone")

    plt.tight_layout()
    plt.show()

    # Perform stationarity test
    check_stationarity(df['Hole Area'], "Hole Area")
    check_stationarity(df['Minimum Ozone'], "Minimum Ozone")

if __name__ == "__main__":
    main()
```
